```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r=3
N=200
x0=0.5
```

```python
def xpiu1(r,x):
    return r*x*(1-x)
```

```python
savenumber=[]
```

```python
savenumber.append(xpiu1(r,x0))
```

```python
savenumber
```

```python
for i in range(N):
        savenumber.append(xpiu1(r,savenumber[i-1]))
```

```python
#savenumber
```

```python
# Plot
plt.plot(savenumber, label='xn')
plt.ylabel('$x_n$')
plt.xlabel('Time')
plt.ylabel
plt.ylim(0,1)
plt.xlim(-1.3,N)
plt.legend()
plt.grid()
```

```python
times=1000
rate=np.linspace(0,4,times)
```

```python
truexn=np.empty((times,10))
```

```python
for p in range(len(rate)):
    initialize=[x0]
    for q in range(N):
        initialize.append(xpiu1(rate[p],initialize[q-1]))
    truexn[p] = [initialize[N-9], initialize[N-8], initialize[N-7], initialize[N-6],initialize[N-5], initialize[N-4], initialize[N-3], initialize[N-2], initialize[N-1], initialize[N]]
```

```python
#truexn
```

```python
x = truexn.reshape(-1)
x
```

```python
ratenew=[]
for i in range(times):
    for p in range(10):
        ratenew.append(rate[i])
```

```python
plt.plot(ratenew,x)
plt.xlabel('grow rate')
plt.ylabel('$x_n$')
plt.ylabel
#plt.legend()
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

interval = (2.8, 4)  # start, end
accuracy = 0.0001
reps = 600  # number of repetitions
numtoplot = 200
lims = np.zeros(reps)

fig, biax = plt.subplots()
fig.set_size_inches(16, 9)

lims[0] = np.random.rand()
for r in np.arange(interval[0], interval[1], accuracy):
    for i in range(reps-1):
        lims[i+1] = r*lims[i]*(1-lims[i])

    biax.plot([r]*numtoplot, lims[reps-numtoplot:], 'k.', markersize=.02)

biax.set(xlabel='grow rate', ylabel='$x_n$', title='logistic map')
plt.show()

# [r]*200 is a way of give to the number 'r' a multiplicity of 200, so it will give 200 quantity of 'r's
# lims[reps-numtoplot:]  will plot all the numbers from a certain point on.
```
