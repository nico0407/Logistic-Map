```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r=3
N=200
x0=0.5
```

```python
def xpiu1(r,x):
    return r*x*(1-x)
```

```python
savenumber=[]
```

```python
savenumber.append(xpiu1(r,x0))
```

```python
savenumber
```

```python
for i in range(N):
        savenumber.append(xpiu1(r,savenumber[i-1]))
```

```python
savenumber
```

```python
# Plot
plt.plot(savenumber, label='xn')
plt.ylabel('$x_n$')
plt.xlabel('Time')
plt.ylabel
plt.ylim(0,1)
plt.xlim(-1.3,N)
plt.legend()
plt.grid()
```

```python
times=1000
rate=np.linspace(0,4,times)
```

```python
truexn=np.empty((times,10))
```

```python
for p in range(len(rate)):
    initialize=[x0]
    for q in range(N):
        initialize.append(xpiu1(rate[p],initialize[q-1]))
    truexn[p] = [initialize[N-9], initialize[N-8], initialize[N-7], initialize[N-6],initialize[N-5], initialize[N-4], initialize[N-3], initialize[N-2], initialize[N-1], initialize[N]]
```

```python
truexn
```

```python
x = truexn.reshape(-1)
x
```

```python
ratenew=[]
for i in range(times):
    for p in range(10):
        ratenew.append(rate[i])
```

```python
ratenew
```

```python
plt.plot(ratenew,x)
plt.xlabel('grow rate')
plt.ylabel('$x_n$')
plt.ylabel
#plt.legend()
plt.grid()
```

```python
plt.plot(ratenew,x)
plt.xlabel('grow rate')
plt.ylabel('$x_n$')
plt.ylabel
#plt.legend()
plt.ylim(0,1)
plt.xlim(2.8,4)
plt.grid()
```
